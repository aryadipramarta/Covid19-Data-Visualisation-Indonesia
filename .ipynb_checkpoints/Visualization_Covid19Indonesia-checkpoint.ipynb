```python
#plot visualisation for Corona Virus based on Daily Case in Province Indonesia (1 Mei 2021)
ax = gplt.choropleth(
    covid_indo,
    hue="Daily_Case",
    edgecolor="white",
    linewidth=2,
    legend=True,
    cmap="OrRd",
    scheme="FisherJenks",
    alpha = 1,
    projection=gcrs.AlbersEqualArea(),
    figsize=(20, 20)
)

gplt.polyplot(
    indonesia_map,
    ax=ax,
    edgecolor="black",
    linewidth=2,
    zorder=2
)

ax.set_title(
    "Peta Sebaran Kasus Positif Harian COVID19 di Indonesia Tanggal 1 Mei 2021 per Provinsi",
    fontsize=34
)
```
